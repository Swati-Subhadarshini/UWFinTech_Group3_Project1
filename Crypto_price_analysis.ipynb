    Base this DataFrame on the NAV prices of the four portfolios and on the closing price of theCryptocurrencies.

```python
# Prepare for the analysis by converting the dataframe of closing prices to daily returns
# Drop any rows with all missing values
# Review the first five rows of the daily returns DataFrame.
crypto_daily_returns = crypto_portfolio_df.pct_change().dropna()
crypto_daily_returns.head()
```

```python
# Calculate and plot the cumulative returns of the 5 cryptocurrencies
# Review the last 5 rows of the cumulative returns DataFrame
crypto_cumulative_returns = (1 + crypto_daily_returns).cumprod()
display(crypto_cumulative_returns.head())
crypto_cumulative_returns.tail()
```

Step 1: Use the Pandas std function to calculate the standard deviation for each of the 5 cryptocurrencies. Review the standard deviation calculations, sorted from smallest to largest.

```python
# Calculate and sort the standard deviation for all 5 crypto portfolios.
# Review the standard deviations sorted smallest to largest
crypto_std_df = crypto_daily_returns.std().sort_values()
crypto_std_df
```

Step 2: Calculate the annualized standard deviation for each of the 5 cryptocurrencies. To do that, multiply the standard deviation by the square root of the number of trading days. Use 252 for that number

```python
# Calculate and sort the annualized standard deviation (252 trading days) of the 5 currencies
# Review the annual standard deviations smallest to largest
##Trading Days 01-01-2021 - 10-09-2021 
trading_days=194
crypto_annual_std = crypto_std_df * np.sqrt(trading_days)
crypto_annual_std.sort_values()
```

Analyze the Risk-Return Profile
To determine the overall risk of an asset or portfolio, quantitative analysts and investment managers consider not only its risk metrics but also its risk-return profile. 
After all, if you have two portfolios that each offer a 10% return but one has less risk, you’d probably invest in the smaller-risk portfolio. 
For this reason, you need to consider the Sharpe ratios for each portfolio.

Step 1: Use the daily return DataFrame to calculate the annualized average return data for the 5 cryptocurrencies.
Use 252 for the number of trading days.

```python
# Calculate the annual average return data for the for fund portfolios and the S&P 500
# Use 252 as the number of trading days in the year
# Review the annual average returns sorted from lowest to highest
#Trading Days 01-01-2021 - 10-09-2021 
trading_days = 194

# Use the Pandas mean function to calculate the average daily return for four portfolios and S&P 500 
annual_average_return = crypto_daily_returns.mean() * trading_days

# Sort the annual average returns from lowest to highest
annual_average_return.sort_values()
```

Step 2: Calculate the Sharpe ratios for the 5 cryptocurrencies. To do that, divide the annualized average return by the annualized standard deviation for each. Review the resulting Sharpe ratios, sorted from lowest to highest.¶

```python
# Calculate the annualized Sharpe Ratios for each of the 4 portfolios and the S&P 500.
# Review the Sharpe ratios sorted lowest to highest
sharpe_ratio = annual_average_return / crypto_annual_std
sharpe_ratio.sort_values()
```

Step 3: Visualize the Sharpe ratios for the Cryptocurrencies in a bar chart. Be sure to include the title parameter, and adjust the figure size if necessary.¶

```python
# Visualize the Sharpe ratios as a bar chart
# Include a title parameter and adjust the figure size
sharpe_ratio.hvplot.bar(
    height=500,
    width=700,
    xlabel='Crypto Coins', 
    ylabel='Sharpe Ratio', 
    title="Sharpe Ratio of Cryptocurrencies from Jan-Oct 2021") 
```
